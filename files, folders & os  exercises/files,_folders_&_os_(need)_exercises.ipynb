Task 1:

```python
import os
import shutil
source_directory = '../os-collaboration-exercise-data/'
convert_dates = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
#print(f"{convert_dates['jan']}-dd-country")

for filename in os.listdir(source_directory):
    root,extension = os.path.splitext(filename)
    old_name = os.path.join(source_directory,filename)
    if not os.path.isfile(old_name):
        continue
    if extension!='.txt':
        continue
    parts = root.split('-')
    st = ''
    for i in parts[1]:
        if i=='_':
            st+='-'
        else:
            st+=i
    numeric_filename = f"{convert_dates[parts[0]]}-{st}.txt"
    new_name = os.path.join(source_directory,numeric_filename)
    shutil.move(old_name,new_name)
```

```python
import glob
def rename_files(directory):
    pattern = os.path.join(directory,'*.txt')
    for file_path in glob.glob(pattern):
        dir_name, file_name = os.path.split(file_path)
        root, extension = os.path.splitext(file_name)
        parts = root.split('-')
        st =''
        for i in parts[1]:
            if i=='_':
                st+='-'
            else:
                st+=i
        new_file_path = os.path.join(dir_name,parts[0]+st+'.txt')
        os.rename(file_path, new_file_path)
        #print(f"Renamed '{file_path}' to '{new_file_path}'")
rename_files(source_directory)
```
