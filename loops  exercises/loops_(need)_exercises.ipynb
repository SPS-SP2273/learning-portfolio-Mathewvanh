```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]
temperatures_farenheit = []
for i in temperatures_celsius:
    temperatures_farenheit.append(int(i*(9/5)+32))
print(temperatures_farenheit)
```

```python
for i in range(1,6):
    print(f"{i} :",end = '\t')
    for j in range(1,6):
        print(f'{i*j}',end = '\t')
    print('\n')
```
