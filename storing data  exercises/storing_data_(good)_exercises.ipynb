```python
import numpy as np
np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])
odd_letters = []
for i in np_array_2d:
    if int(i[0])%2!=0:
        odd_letters.append(i[1])
print(odd_letters)
```
