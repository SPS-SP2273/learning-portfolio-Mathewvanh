Task 1:

```python
from matplotlib import pyplot as plt
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = 0.25

# Plotting the error bar plot
plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black', capsize=5,
             label='Y values with errors')

# Adding labels and title
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')

# Grid and legend
plt.grid(alpha=0.25)
plt.legend(loc='lower right')

# Show the plot
plt.show()
```

```python
import numpy as np
x = np.linspace(0,50,200)
y = (np.exp(-x/10))*np.sin(x)
plt.plot(x,y)
plt.xlabel("x",fontsize=15,color = 'darkblue')
plt.ylabel("y",fontsize=15, color = 'darkblue')
plt.grid(alpha = 0.25)
plt.title("$y = e^{-x/10}sin(x)$",fontsize = 20)
plt.ylim(-1,1)
```
