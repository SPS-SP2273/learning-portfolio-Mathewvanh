```python
def calculate_bmi(weight,height):
    if height==0:
        raise ZeroDivisionError
    BMI = weight/(height**2)
    if BMI<18.5:
        return "Underweight"
    elif BMI>=18.5 and BMI<25:
        return "Normal Weight"
    elif BMI>=25 and BMI<30:
        return "Overweight"
    elif BMI>=30:
        return "Obesity"
print(calculate_bmi(90,1.85))
```
