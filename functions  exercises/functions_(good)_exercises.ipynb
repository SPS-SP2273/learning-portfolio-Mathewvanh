```python
def convert_fahrenheit(temperature_farenheit,target_scale = 'Celsius'):
    if target_scale=='Kelvin':
        return (temperature_farenheit-32)*5/9-273.15
    return (temperature_farenheit-32)*5/9
print(convert_fahrenheit(557.6,'Kelvin'))
```
