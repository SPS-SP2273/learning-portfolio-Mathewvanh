```python
def convert(x):
    if x>=70 and x<=100:
        return 'A'
    elif x<70 and x>=50:
        return 'B'
    elif x<50 and x>=35:
        return 'C'
    elif x<35 and x>=0:
        return 'Fail'
    else:
        raise ValueError("Not a valid score!")
```

```python
import ipywidgets as widgets
#print("The student's letter grade is: "+ mark(student_mark))                  
mark_input = widgets.IntText(
    value=input(),
    description="Enter the student's score:",
    disabled=False
)
#convert(mark_input.value)
print(f"The student's letter grade is: {convert(mark_input.value)}")                 
```
